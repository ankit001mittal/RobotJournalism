```python
def search_items(item_number):
    search_string = str(df.Media_Name[item_number])
    search_results = wikipedia.search(search_string)
    relevant_result = search_results[0:1]
    return relevant_result

relevant_list =[]
for i in range(0,len(df)):
    relevant_list.append(search_items(i))
    i+=1
#print(relevant_list)
```

```python
# create a list of all the names you think/know the section might be called
content = ['Plot','Synopsis','Plot synopsis','Plot summary',
             'Story','Plotline','The Beginning','Summary',
            'Content','Premise']

possibles_edit = [i + 'Edit' for i in content]
#then merge those two lists together
all_content = content + possibles_edit

meta_data=['Directed by','Produced by','Screenplay by','Story by','Starring',
           'Music by','Cinematography','Edited by','Distributed by',
           'Release date','Running time','Country','Language','Budget','Box office']

movie_details = pd.DataFrame(index=test_list,columns = all_content)
#movie_details.iloc[0:,0:]
```

```python
for i in test_list:
# load the page once and save it as a variable, otherwise it will request
# the page every time.
# always do a try, except when pulling from the API, in case it gets confused
# by the tttle.
    try:
        wik = wikipedia.WikipediaPage(i)
        r = requests.get(wik.url)
        wt = wtp.parse(r.text)
        soup = bs4.BeautifulSoup(r.text, "html.parser")
        metadata_dict = {}
        table_content = soup.find("table", attrs={"class": "infobox vevent"})
        tbody = table_content.find('tbody')
        trs = tbody.find_all('tr')
        for tr in trs:
            ths = tr.find_all('th')
            tds = tr.find_all('td')
            if tds and ths:
                for th, td in zip(ths, tds):
                    # print(th,td)
                    #print("{}: {}".format(th.text, td.text))
                    metadata_dict[th.text] = td.text
            
    except:
        wik = np.NaN

# a new try, except for the plot

    try:
        for sections in all_content:
            # if that section does exist, i.e. it doesn't return 'None'
            if wik.section(sections) != None:
                # then that's what the plot is! Otherwise try the next one!
                movie_details.loc[i, sections] = wik.section(sections)

        for data in meta_data:
            if data in metadata_dict:
                movie_details.loc[i, data] = metadata_dict[data]

        if wik.summary != None:
            movie_details.loc[i, 'Summary'] = wik.summary

        # if none of those work, or if the page didn't load from above, then plot
        # equals np.NaN
    except:
        plot= np.NaN
```
